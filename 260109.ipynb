```python
c= [1,2,3,5]

result = []
for i in c:
    result.append(i * 2)

print(result)
```

```python
c= [1,2,3,5]
d = [i*2 for i in c]
print(d)
```

```python
d= [[1,2], [3,4], [5,6]]
result_1 = []
result_2 = []

#1
for i in d:
    result_1.append(i[0] + i[1])
print(result_1)

#2
for i, j in d:
    result_2.append(i+j)
print(result_2)
```

```python
c= [1,2,3,5]
result = []

for i in c:
    if i % 2 != 0:
        result.append(i)

print(result)
```
